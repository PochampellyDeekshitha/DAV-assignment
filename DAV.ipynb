```python
#  Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#  Optional: Use a clean theme
sns.set(style="whitegrid")

#  Load CSV file (adjust filename if different)
df = pd.read_csv("car_prices.csv")

#  Clean column names
df.columns = df.columns.str.strip().str.lower()

#  Drop rows with missing important data
df = df.dropna(subset=["sellingprice", "make", "year"])

#  Ensure price is numeric
df["sellingprice"] = pd.to_numeric(df["sellingprice"], errors="coerce")
df = df.dropna(subset=["sellingprice"])

#  1. Average Selling Price
avg_price = np.round(df["sellingprice"].mean(), 2)
print(f"Average car price: ₹{avg_price}")

#  2. Most Popular Brand
top_brand = df["make"].value_counts().idxmax()
print(f"Most popular brand: {top_brand}")

#  3. Average Price by Year
avg_price_per_year = df.groupby("year")["sellingprice"].mean().round(2)
print("\nAverage price per year:")
print(avg_price_per_year.head())

#  Selling Price Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["sellingprice"], bins=50, kde=True, color='skyblue')
plt.title("Car Selling Price Distribution")
plt.xlabel("Selling Price")
plt.ylabel("Frequency")
plt.show()

#  Top 10 Car Brands by Count (Fixed Warning)
plt.figure(figsize=(10, 5))
top_makes = df["make"].value_counts().head(10)
sns.barplot(x=top_makes.index, y=top_makes.values, hue=top_makes.index, palette="viridis", legend=False)
plt.title("Top 10 Car Brands by Count")
plt.xlabel("Car Brand")
plt.ylabel("Number of Listings")
plt.xticks(rotation=45)
plt.show()

#  Average Price by Year
plt.figure(figsize=(10, 5))
avg_price_per_year.plot(kind="line", marker='o', color='orange')
plt.title("Average Selling Price by Year")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()
```

```python
#  Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Clean style
sns.set(style="whitegrid")

#  Load dataset (change filename if needed)
df = pd.read_csv("car_prices.csv")

#  Check basic info
print(df.head())
print(df.columns)

#  Clean column names
df.columns = df.columns.str.strip().str.lower()

#  Drop missing values in key columns
df = df.dropna(subset=["sellingprice", "make", "year"])
df["sellingprice"] = pd.to_numeric(df["sellingprice"], errors="coerce")
df = df.dropna(subset=["sellingprice"])

#  NumPy & Pandas operations

# 1 Average selling price
avg_price = np.round(df["sellingprice"].mean(), 2)
print(f"Average selling price: ₹{avg_price}")

# 2 Most popular car brand
top_brand = df["make"].value_counts().idxmax()
print(f"Most popular car brand: {top_brand}")

# 3 Average price by year
avg_price_per_year = df.groupby("year")["sellingprice"].mean().round(2)
print("Average price by year:")
print(avg_price_per_year)

#  Visualization 1: Price Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["sellingprice"], bins=50, kde=True, color="skyblue")
plt.title("Selling Price Distribution")
plt.xlabel("Selling Price")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

#  Visualization 2: Top 10 Car Brands
plt.figure(figsize=(10, 5))
top_makes = df["make"].value_counts().head(10)
sns.barplot(x=top_makes.index, y=top_makes.values, hue=top_makes.index, palette="Set2", legend=False)
plt.title("Top 10 Car Brands by Listings")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#  Visualization 3: Average Price Over Years
plt.figure(figsize=(10, 5))
avg_price_per_year.plot(kind="line", marker='o', color='green')
plt.title("Average Car Price Over the Years")
plt.xlabel("Year")
plt.ylabel("Avg Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```
